## Unsupervised learning

### Books recommendation based on wish list:

```python
# Creating pivot table
matrix = df_ratings_c.pivot_table(index='book_id', columns='user_id', values='rating')

# fill NA values with 0 and make the index into a column
matrix = matrix.fillna(0).reset_index()
```

```python
matrix.head()
```

```python
# Slice matrix to only include ratings
x_cols = matrix[matrix.columns[1:]]
x_cols.head()
```

**PCA**

```python
##### Import PCA
from sklearn.decomposition import PCA

# Create PCA instance
model_pca = PCA(n_components=10)

# Apply the fit_transform method
pca_features = model_pca.fit_transform(x_cols)
```

```python
pca_features.shape
```

```python
# Finding intrinsec dimension

features = range(model_pca.n_components_)
```

```python
plt.bar(features, model_pca.explained_variance_)
plt.xticks(features)
plt.ylabel('variance')
plt.xlabel('PCA feature')
plt.show()
```

Intrinsic dimension assumption: 5 features

**TSNE**

```python
# Import TSNE
from sklearn.manifold import TSNE

# Create a TSNE instance: model
model = TSNE(learning_rate=300)

# Apply fit_transform to samples: tsne_features
tsne_features = model.fit_transform(pca_features)

# Select the 0th feature: xs
xs = tsne_features[:,0]

# Select the 1st feature: ys
ys = tsne_features[:,1]
```

```python
# Scatter plot:
plt.scatter(xs, ys, alpha=0.5)
plt.show()
```
