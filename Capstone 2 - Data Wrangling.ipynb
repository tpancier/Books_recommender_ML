# DATA WRANGLING

**Read files, parse data, create dataframes**

```python
# Reading, parsing XML file, saving as nested dictionary

import xml.etree.ElementTree as ET
import pandas as pd

dic_xml = {}
dic_aut = {}
dic_work = {}
files_to_read = 33288638

for nr_file in range(1,files_to_read+1):
    file = 'books_xml/' + str(nr_file) + '.xml'
    try:
        tree = ET.parse(file)
        root = tree.getroot()
        dic_xml[nr_file] = {}
        dic_aut[nr_file] = {}
        dic_work[nr_file] = {}

        # creating books dictionary
        for i in range(len(root[1])):
            child = root[1][i].tag
            element = root[1][i].text
            dic_xml[nr_file].update({child:element})
        
        # creating authors dictionary
        for i in range(len(root[1][26][0])):
            child = root[1][26][0][i].tag
            element = root[1][26][0][i].text
            dic_aut[nr_file].update({child:element})
            
        # creating work id dictionary
        for i in range(len(root[1][17])):
            child = root[1][17][i].tag
            element = root[1][17][i].text
            dic_work[nr_file].update({child:element})
            
    except Exception:
        pass
```

```python
len(root[1][17])
```

```python
# Creating books dataframe

df_books = pd.DataFrame.from_dict(dic_xml, orient='index')

df_books.head()
```

```python
# Creating authors dataframe

df_author = pd.DataFrame.from_dict(dic_aut, orient='index')

df_author.head()
```
